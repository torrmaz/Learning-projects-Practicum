# Шаг 4 Формулирование и проверка статистических гипотез:

Для проверки гипотезы о различии дохода между клиентами, которые ушли и которые остались, мы можем использовать t-критерий для независимых выборок. 
Сформулируем статистическую гипотезу о равенстве двух средних I:
- Нулевая гипотеза (H0): Средний доход клиентов, которые ушли из банка, равен среднему доходу клиентов, которые остались.
- Альтернативная гипотеза (H1): Средний доход клиентов, которые ушли из банка, отличается от среднего дохода клиентов, которые остались.

Далее проведем t-критерий для независимых выборок, чтобы оценить статистическую значимость различий. Если p-value окажется меньше выбранного уровня значимости (обычно 0.05), то мы можем отклонить нулевую гипотезу и заключить, что средний доход между клиентами, которые ушли и которые остались, различается.


Для проверки гипотезы о различии количества баллов собственности между клиентами, которые покинули банк, и клиентами, которые остались, мы можем также использовать t-критерий для независимых выборок.


Сформулируем статистическую гипотезу о равенстве двух средних II:

- Нулевая гипотеза (H0): Среднее количество баллов собственности у оставшихся клиентов равно среднему количеству баллов собственности у ушедших клиентов.
- Альтернативная гипотеза (H1): Среднее количество баллов собственности у оставшихся и ушедших клиентов отличается.

После проведения анализа и оценки значимости различий, можно сделать вывод о том, действительно ли количество баллов собственности отличается у клиентов, которые покинули банк, и тех, которые остались.

Если данные не имеют нормальное распределение, придется использовать непараметрические методы, такие как критерий Манна-Уитни или критерий Уилкоксона.

Сформулируем статистическую гипотезу  III:

- Нулевая гипотеза (H0):  Наличие кредитной карты не влияет на вероятность оттока.
- Альтернативная гипотеза (H1):  Клиенты с кредитной картой имеют меньшую вероятность оттока.
долю оттока клиентов в каждой из этих групп мы выяснили на прошлом шаге, будем использовать z-критерий , для сравнения долей оттока в этих двух группах.

### Обработка пропущенных значений

Пропуски содержаться в столбцах age и balance

```python
# Для оценки типа пропусков рассмотим коррелируют ли они с другими признаками\
# Создадим копии исследуеимых столбцов 
df['age_null']=df['age']
df['balance_null']=df['balance']
#cоздадим новый признак, принимающий 1, если в наблюдении пропущено значение признака, и 0, если в наблюдении значение признака не пропущено
df['age_null'] = df['age_null'].apply(lambda x: 1 if pd.isna(x) else 0)
df['balance_null'] = df['balance_null'].apply(lambda x: 1 if pd.isna(x) else 0)
#Выведем корелляциии на экран
display('Корреляция balance_null')
display(df.corrwith(df['balance_null']).sort_values(ascending=True).apply('{:.2f}'.format))
display('Корреляция age_null')
display(df.corrwith(df['age_null']).sort_values(ascending=True).apply('{:.2f}'.format))

display(df.groupby('equity')['balance_null'].sum())
display(df.groupby('city')['age_null'].sum())
```

Значения нового столбца balance_null сильно коррелируют с даными в столбцах equity и products, природа пропуска не случайна (Missing Not At Random) без дополнительного обоснования их нельзя отбрасывать или заполнять одним значением, т.к. это приведёт к заметным искажениям

Значения столбца age_null с другими столбцами коррелирует слабо мы не можем это распределение назвать случайным в рамках каждой из групп(Missing Completely At Random / Отсутствует совершенно случайно) Заполним Заполним медианным значением, т.к пропусков менее 1%

Согласно интервалам с оттоком выше чем по банку и ниже чем по банку разделим клиентов на возрастные категории.

```python
# функция age_gr для распределения клиентов по возрастным группам:
def age_gr(age):
    if 24 > age >= 18:
        return '18-24'
    if  35>= age >= 25:
        return '25-35'
    if 49 >= age >= 36:
        return '36-49'
    if 60 >= age >= 50:
        return '50-60'
    if 61 <= age:
        return '61+'  
df['age_group'] = df['age'].apply (age_gr)
```

## Шаг  Формулирование и проверка статистических гипотез:

### Гипотеза I Влияние дохода клиентов на отток
**Нулевая гипотеза (H0)**: Средний доход клиентов, которые ушли из банка, равен среднему доходу клиентов, которые остались.

**Альтернативная гипотеза (H1)**: Средний доход клиентов, которые ушли из банка, отличается от среднего дохода клиентов, которые остались.

Чтобы понять, можно ли применять t-тест, нужно проверить три условия:

1. Являются ли две выборки независимыми.

2. Соответствуют ли элементы данных в выборках какому-либо нормальному распределению.

3. Имеют ли данные две выборки схожие дисперсии

При невыполнении этих условий будем использовать непараметрический тест Манна-Уитни 

```python
# найдем верхнюю границу q3+1.5*IQR
q3 = df['est_salary'].quantile(0.75)
iqr = q3 - df['est_salary'].quantile(0.25)
upper_bound = q3 + (1.5 * iqr)
print(upper_bound)
# Отсекаем выбросы по этой границе
df_h_1=df[~(df['est_salary'] > upper_bound)]
a = df_h_1[df_h_1['churn'] == 1]['est_salary'].head(1700)
b = df_h_1[df_h_1['churn'] == 0]['est_salary'].head(1700)



fig, axes = plt.subplots(1,2,squeeze=False, figsize=(15, 6))
plt.title('est_salary без вбросов')
sns.histplot(data=df[df['est_salary']<=upper_bound],bins=60, x = 'est_salary', hue='churn', stat='density',  kde = True,common_norm=False,ax=axes[0][0]);
axes[0][0].set_title('Нормированая гистограма в разрезе churn')
axes[0][0].legend([ 'Отточные','Не отточные'])
sns.boxplot(x = df_h_1.est_salary,ax=axes[0][1]);
a.size
```

```python
def normality(x,y):
    # проведём тест на соотвествие данных о доходах отточных клиентов нормальному распределеню
    alpha = 0.05  # критический уровень статистической значимости
    shapiro_test_stat, p_value = st.shapiro(x)
    print('p-значение: ', p_value)
    if p_value < alpha:
        print('Распределение не нормально')
    else:
        print('Распределение нормально')  
            # проведём тест на соотвествие данных о доходах отточных клиентов нормальному распределеню
    alpha = 0.05  # критический уровень статистической значимости
    shapiro_test_stat, p_value = st.shapiro(y)
    print('p-значение: ', p_value)
    if p_value < alpha:
        print('Распределение не нормально')
    else:
        print('Распределение нормально')  
```

```python
normality(a,b)
```

Не можем применить тест пирсона или спирмена ко всем данным. Применили частично учитывая размер выборок
Выборки независимы.

Распределение значений не является нормальным. Выборочное распределение выборочного среднего является приблизительно нормальным, если размер выборки достаточно велик, даже если распределение генеральной совокупности не является нормальным

То есть мне совсем не важно как распределены данные в выборках, если их размер достаточно велик.Ведь ttest работает с выборочными средними.

Дисперсии не равны.При неравенстве дисперсий используется поправка Уэлча с параметром `equal_var=False`, разность дисперсий не является препятствием использовать `ttest`.

Выборки независимы, данные в выборках распределены не нормально, дисперсии различаются.

Вбросы остались даже после отсечения по верхней границе усов 1,5*IQR, это может искажать результаты параметрических тестов. более уместным будет использование непараметрического теста - теста Манна-Уитни. Он не требует предположения о нормальности распределения данных и может быть применен даже при наличии выбросов и различающихся дисперсий.

```python
# непараметрический тест Манна-Уитни. u-test

alpha = 0.05  

results = st.mannwhitneyu(a, b)

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу:Средний доход клиентов, которые ушли из банка, отличается от среднего дохода клиентов, которые остались.")
else:
    print("Не получилось отвергнуть нулевую гипотезу:Средний доход клиентов, которые ушли из банка, равен среднему доходу клиентов, которые остались.")
```

### Гипотеза II Балл кредитного скорринга у отточных клиентов не отличается от неотточных.

**Нулевая гипотеза (H0)**: Балл кредитного скорринга у отточных клиентов не отличается от неотточных.

**Альтернативная гипотеза (H1)**: Балл кредитного скоринга отточных отличается от неотточных.

Чтобы понять, можно ли применять t-тест, нужно проверить три условия:

1. Являются ли две выборки независимыми.

2. Соответствуют ли элементы данных в выборках какому-либо нормальному распределению.

3. Имеют ли данные две выборки схожие дисперсии

При невыполнении этих условий будем использовать непараметрический тест Манна-Уитни 

```python
a = df[df['churn'] == 1]['score']
b = df[df['churn'] == 0]['score']


hist_by_feature('score',80,'Нормированая гистограма в разрезе churn')
```

```python
# оценим дисперсии
print(a.var())
print(b.var())
print('Дисперсии не равны')
```

```python
a=a.head(1500)
b=b.head(1500)
stat, p = pearsonr(a, b)
print('stat-%.3f, p-%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

Выборки независимы, данные в выборках не похожи на нормально распределенные, дисперсии различаются.
Изначально мы указали, что  t-тест нельзя использовать, так как данные не имеют нормальное распределение. 

Однако, с учетом Центральной Предельной Теоремы, можно сделать вывод, что при достаточно большом объеме выборки (обычно больше 30 наблюдений у нас 1700), распределение средних значений выборок стремится к нормальному распределению независимо от формы распределения исходной выборки. 

Таким образом, если объем выборки достаточно большой, можно применять t-тест даже в случае, когда данные изначально не имеют нормальное распределение. В данном случае, используя ЦПТ, мы можем обосновать применение t-теста для проведения статистических проверок гипотез.

Используем t-критерий Стьюдента для проверки статистической гипотезы

```python
# Проведение t-теста для независимых выборок
t_stat, p_val = st.ttest_ind(a, b,equal_var=False)

# Вывод результатов
print("t-statistic:", t_stat)
print("p-value:", p_val)
if p_val < 0.05:
    print("Отвергаем нулевую гипотезу:Балл кредитного скоринга отточных отличается от неотточных.")
else:
    print("Не получилось отвергнуть нулевую гипотезу:Балл кредитного скорринга у отточных клиентов не отличается от неотточных")
```

### Гипотеза III Клиенты с кредитной картой менее склонны к оттоку.

**Нулевая гипотеза (H0)**: Наличие кредитной карты не влияет на вероятность оттока.

**Альтернативная гипотеза (H1)**: Клиенты с кредитной картой имеют меньшую вероятность оттока.


```python
a = df[df['credit_card'] == 1]['churn']
b = df[df['credit_card'] == 0]['churn']
```

Для гипотез, связанных с проверкой бинарных признаков, нужен тест пропорций. Выполним левосторонний z-test. будем сравнивать наши z значения с общим оттоком по банку churn_val

```python
# проведем z-test
# критический уровень статистической значимости
alpha = 0.05

# уровень оттока по банку
churn_val = df['churn'].mean()

#количество общее
nobs=a.size
# количество успешных
count =a.sum()

# z-test
z_stat, pval = proportions_ztest(count, nobs, churn_val, alternative= 'smaller')
print('p-значение:', '{0:0.3f}'.format(pval))
if pval < alpha:
    print('Отвергаем нулевую гипотезу:Клиенты с кредитной картой имеют меньшую вероятность оттока')
else:
    print('Не получилось отвергнуть нулевую гипотезу:Наличие кредитной карты не влияет на вероятность оттока.')
```
