## Предобработка данных

### Обработаем столбец is_apartment 

```python
display('Пропусков в столбце ',data['is_apartment'].isna().sum()) # поиск пропусков 20924
data['is_apartment'] = data['is_apartment'].fillna(False).astype('bool')# замена пропусков на false + bool type
data['is_apartment'].value_counts()
```

Меняем тип на bool опираясь на описание данных,заодно заменяем все пропуски столбца значением False удобства последующего анализа.

### Рассмотрим столбцы  cityCenters_nearest,    parks_around3000,   ponds_around3000.

```python
data=data.rename(columns={'cityCenters_nearest':'city_centers_nearest',
                          'parks_around3000':'parks_around',
                          'ponds_around3000':'ponds_around'})# переименование столбцов
data.columns
```

 Переименовываем столбцы в соответствии со змеиным регистром

### Приведем столбец  first_day_exposition к формату datetime64

```python
# преобразуем first_day_exposition к типу datetime с помощью pd.to_datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

### Столбец locality_name
Избавляемся от неявных дубликатов заменяя 'ё' на 'е',а 'городской поселок' на 'поселок городского типа'
Заодно проверяем на явные дубликаты

```python
display('Уникальных названий', data['locality_name'].nunique())
display(data['locality_name'].unique())#уникальные значения'locality_name'

display(data['locality_name'].isna().sum())#49 пропусков в названии населенных пунктов
data = data.dropna(subset=['locality_name']) #удаление пропусков

data['locality_name'] = data['locality_name'].str.lower()
data['locality_name'] =data['locality_name'].str.replace('ё', 'е').str.replace('городской поселок', 'поселок городского типа', regex=True)

display('Уникальных названий', data['locality_name'].nunique())
```

```python
#колличество пропусков 9160
display('Колличество пропусков',data.ceiling_height.isna().sum()) 
#краткая информация по столбцу "ceiling_height"
display(data['ceiling_height'].describe()) 
#Все значения больше 10 делим на 20 и возвращаем на свои места
data.loc[data['ceiling_height']>=10, 'ceiling_height'] = data['ceiling_height']/10
display('колличество квартир с потолками равными')
display(data['ceiling_height'].value_counts()) 

data['ceiling_height'].sort_values().plot(
    y='ceiling_height', 
    kind='hist', 
    bins=50, 
    range=(2,6),
    grid=True);
```

### Добавим тип этажа квартиры

```python
def floor (row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['floor_type'] = data.apply(floor, axis=1)#применение функции ко всем строкам 
data.value_counts('floor_type')
```

```python
def issl(wer,bool):    
    if bool==0:
        display(data[wer].describe())
        #Диаграмма размаха
        data.boxplot(wer, figsize=(10,12), grid=True)
        plt.title(f'Диаграмма размаха {wer}')
        #Гистограмма total area
        data.plot(y = wer, kind = 'hist', bins = 100,title=f'Распределение предложений квартир по {wer}', grid=True, figsize = (10,5), range = (0,data[wer].max()))
        plt.xlabel(f'{wer}')
        plt.ylabel('Количество предложений')
        plt.show()
        # найдем верхнюю границу q3+1.5*IQR
        upper_bound = data[wer].quantile(0.75) + (1.5 *(data[wer].quantile(0.75)-data[wer].quantile(0.25)))
        #Гистограмма total area без вбросов
        data_area=data.query(f'{wer} <= @upper_bound')
        data_area.plot(kind='hist', y=wer,title=f'Распределение предложений квартир по {wer}',bins=100, figsize=(15,5), grid=True, range = (data_area[wer].min(),data_area[wer].max()))
        plt.xlabel(f'{wer}')
        plt.ylabel('Количество предложений')
        plt.show()
    else:
        upper_bound = data[wer].quantile(0.75) + (1.5 *(data[wer].quantile(0.75)-data[wer].quantile(0.25)))
        #Гистограмма total area без вбросов
        data_area=data.query(f'{wer} <= @upper_bound')
        return data_area
```
